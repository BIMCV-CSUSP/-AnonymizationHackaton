# Exercise 2: De-identification of Dicom tags with CTP

### The purpose of this exercise is to learn how to install and use basically the open-source software CTP from RNSA to de-identify the Personal Health Information (PHI) inside the tags of DICOM files in a local device.

## Documentation

Software CTP manual webpage: 
http://mircwiki.rsna.org/index.php?title=CTP-The_RSNA_Clinical_Trial_Processor


## Input data of the exercise

Two DICOM files from a conventional radiography session whose tags han been modified.

To get these files, go to the link: [input files DICOM](https://drive.google.com/drive/folders/11b_zpT8lg1QGqe3kK2OLqLqUKtBc81w-?usp=sharing)


## Before CTP instalation

1. Java 1.7 (or better) JRE is required, Java 1.8 JRE recommended.
2. For Windows, java must be of 32 bits. 
3. To download software CTP, go to webpage: http://mirc.rsna.org/query
4. Go to the left menu and click on ‘Download Software’.
5. Download the file 'CTP-installer.jar'.



### Instalation of CTP

Execute the file on terminal:
```
java -jar CTP-installer.jar
```
Choose the folder to put the program. Next, choose the port number for the CTP server that will give a offline monitoring webpage.

Once the instalation process is finished and the java window is closed, the 'CTP' folder is created and a new java window will be opened to start the configuration of the server. Meanwhile, a support webpage will show up to help in the configuration:

http://mircwiki.rsna.org/index.php?title=The_CTP_Launcher_Configuration_Editor

### Configuration of CTP

Leave the default configuration of the 'server', all the exercise will be offline.

It is necessary to create a 'pipeline'. To do that, go to 'Pipeline' option on the top menu and click the option 'New Pipeline' or using the several templates. For this exercise, choose the template'CTP Classroom', click in its name on the lateral left menu and change the variable 'name' on the right.

**name:**``` CTP_Exercise_Hackathon```

A CTP pipeline structure has three basic components:

*   Import the files
*   One or more process (like Anonymization)
*   Export the files

Because this exercise is offline, we need to change te Import step in this template. Choose 'DicomImportService' on the lateral left panel. Go to the top menu and click the option 'Edit' and choose 'Remove'.

Following, let's add the new Import, go to'ImportService' and choose the option 'DirectoryImportService'. This process has to be on first place, so drag it with the mouse at the top of the pipeline list of the left panel.

Check the default values and be sure every process has port number and they are not similar.

**A warning for Linux:** the port numbers of every pipeline process has to be more than 1024.

When the pipeline configuration is finished, save it clicking on 'File' and choose 'Save'.

Go to the tap 'General' and push the button 'Start' to active the CTP server.

If you close the java window, you can come back if you go to de 'CTP' folder and execute in a terminal:

```
java -jar Launcher.jar
```

## Monitoring webpage access
On 'General' tap, push the button 'CTP Home Page' and a offline webpage will be opened for monitoring the CTP. As this is the first time, log in username 'admin' and password 'password'.

The administrator access let you more options on the left panel menu. You can see the 'System Summary' on the right panel, look that the pipeline created is on the list. The summary table notifies the number of images that are in 'Import' and 'Export' stage, and also the images with some process problem that are in 'Quarantine' stage.


## Modify DICOM Anonymizer Script

It is recommended to copy the file 'DicomAnonymizer.script' inside the folder 'scripts' in 'CTP' directory.

### Anonymizer Script access by monitoring webpage
As administrator, go to the left panel and click on 'DicomAnonymizer' of the pipeline. Following, click on 'Edit the Anonymizer Script' on the right panel.

A table will show up with all the DICOM tag that will be important in an anonymization process. At 'Select' column, you can select the tags.

At the top, there is the menu to configure the Anonymizer script. The 'Help' option goes to useful help webpages.

This webpage explains the functions of the script:
http://mircwiki.rsna.org/index.php?title=The_CTP_DICOM_Anonymizer

This webpage explains the configuration menu:
http://mircwiki.rsna.org/index.php?title=The_CTP_DICOM_Anonymizer_Configurator

### Choosing confidentiality profile

Go to 'Profiles' of the configuration menu, put the mouse on 'DICOM' option and an panel will be opened where you can choose the possible confidentiality profiles. Choose one and save the script at 'File' on the top menu or the disquette icon on the right top.

## Placing the input files
Copy the input files on the path indicated on the variable 'import' of the pipeline process 'Directory Import Service'.
If you did't make any change in the CTP configuration, the path within the 'CTP' folder is probably:
```
roots/DirectoryImportService/import
```
When the CTP server is working, those files will be deleted in this folder and you can check them in the path indicated on the variable 'root' of the first 'FileStorageSystem' with different filenames.



## Monitoring the pipeline
Go to the 'Home CTP Page', and log in as administrator.

Click on the pipeline name on the left panel menu and see the summary of the processes.

Click on 'FileStorageService-Anonymized' and click on 'View Stored Studies for __default' on the right panel. A table with the sessions will show up of the anonymized DICOM images.

Click on 'List' and a list of all the DICOM files will show up. The 'List' option of every file is the access of the DICOM tag anonymized values.
